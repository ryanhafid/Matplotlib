```python
angles=np.linspace(0,2*np.pi,len(subjects), endpoint=False)
print(angles)
```

```python
angles=np.concatenate((angles,[angles[0]]))
print(angles)
```

```python
subjects.append(subjects[0])
alice.append(alice[0])
veihor.append(veihor[0])
horiver.append(horiver[0])
horihor.append(horihor[0])
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)
ax.plot(angles,alice)
ax.fill(angles, alice, alpha=0.25, color='g')
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(polar=True)
#basic plot
ax.plot(angles,alice, 'o--', color='g', label='Veveiveri')
#fill plot
ax.fill(angles, alice, alpha=0.25, color='g')
#Add labels
ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)


plt.tight_layout()
plt.legend()
plt.show()
```
